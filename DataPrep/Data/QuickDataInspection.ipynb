```python
import numpy as np
import pandas as pd
```

```python
cities_age = pd.read_csv('CITIES_AGE.csv',sep="|")
cities_coord = pd.read_csv('CITIES_COORD.csv',sep="|")
cities_economy = pd.read_csv('CITIES_ECONOMY.csv',sep="|")
cities_env = pd.read_csv('CITIES_ENVIRONMENT.csv',sep="|")
cities_labour = pd.read_csv('CITIES_LABOUR.csv',sep="|")
cities_population = pd.read_csv('CITIES_POPULATION.csv',sep="|")
cities_territory = pd.read_csv('CITIES_TERRITORY.csv',sep="|")
```

```python
country_cities = cities_economy[['METRO_ID','Metropolitan areas']]
country_cities = country_cities[cities_economy['Year']==2001]
country_cities
```

```python
#function that selects first few letters of a string

def first_letters(string):
    output = ''
    for char in range(len(string)):
        if ord(str(string[char])) >= 65:
            output += str(string[char])
    return output
```

```python
#create dictionairy of countries and add cities to them

metro_id_list = list(country_cities['METRO_ID'])
country_initials_list = []
for country in metro_id_list:
    country_initials_list.append(first_letters(country))

country_initials_set = set(country_initials_list)
country_initials_set

country_dict = {}

for init in country_initials_set:
    country_dict[init] = set(())
    for index, row in country_cities.iterrows():
        if init in row['METRO_ID']:
            country_dict[init].add(row['Metropolitan areas'])
            
print(country_dict)
    
```

```python
pd.DataFrame.from_dict(country_dict, orient='index')
```
